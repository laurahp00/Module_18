```python
import os
from os import chdir
from glob import glob
import pandas as pd
import csv 
import numpy as np
```

```python
# Create a DataFrame
file1 = "JC-202010-citibike-tripdata.csv"
file2 = "JC-202011-citibike-tripdata.csv"
file3 = "JC-202012-citibike-tripdata.csv"

october_df = pd.read_csv(file1)
november_df = pd.read_csv(file2)
december_df = pd.read_csv(file3)

october_df
```

```python
# Check the datatypes
october_df.dtypes
```

```python
november_df.dtypes
```

```python
# Tripduration as delta, start and end as datetimes
october_df['tripduration'] = pd.to_timedelta(october_df['tripduration'], unit='s')
october_df['starttime'] = pd.to_datetime(october_df['starttime'])
october_df['stoptime'] = pd.to_datetime(october_df['stoptime'])

november_df['tripduration'] = pd.to_timedelta(november_df['tripduration'], unit='s')
november_df['starttime'] = pd.to_datetime(november_df['starttime'])
november_df['stoptime'] = pd.to_datetime(november_df['stoptime'])

december_df['tripduration'] = pd.to_timedelta(december_df['tripduration'], unit='s')
december_df['starttime'] = pd.to_datetime(december_df['starttime'])
december_df['stoptime'] = pd.to_datetime(december_df['stoptime'])
```

```python
# Check the datatypes again

#october_df.dtypes
#november_df.dtypes
december_df.dtypes
```

```python
# Gender needs to be changed from int
# I looked it up, for gender it is (Zero=unknown; 1=male; 2=female);

gender_mapping = {0: 'Unknown', 1: 'Male', 2: 'Female'}
october_df['gender'] = october_df['gender'].replace(gender_mapping)
november_df['gender'] = november_df['gender'].replace(gender_mapping)
december_df['gender'] = december_df['gender'].replace(gender_mapping)
```
